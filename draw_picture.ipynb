```python
times = []
with open('runlog.txt') as f:
    line = f.readline()
    while line:
        if(line[0]=='t'):
            print(line[13:-1])
            times.append(float(line[13:-1]))
        line = f.readline()
```

```python
new_times = []
for i in range(len(times)):
    time_all=0
    time_all+=times[i]
    
    if i % 4 == 3:
        new_times.append(time_all)
        
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif']=['SimHei']
data = new_times
plt.hist(data, bins=30,edgecolor='black') # 使用plt.hist函数绘制直方图
plt.xlabel('处理时间/s') # 设置x轴标签
plt.ylabel('出现频率') # 设置y轴标签
plt.title('景佳微官方测试数据集压缩效果图') # 设置标题
plt.show() # 显示图像
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# 定义参数
image_path = "ARGB_GT/"  # 指定图片路径
image_names = ["{}.png".format(i) for i in range(9, 18)]  # 图片名称列表

# 加载图片并拼接
images = [Image.open(image_path + name) for name in image_names]
width, height = images[0].size
total_width = 3 * width
total_height = 3 * height
new_image = Image.new('RGB', (total_width, total_height))
for i in range(len(images)):
    row = int(i / 3)
    col = i % 3
    new_image.paste(images[i], (col * width, row * height))

# 显示拼接后的图片
plt.imshow(new_image)
plt.show()
```
